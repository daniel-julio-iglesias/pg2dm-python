```python
def load_or_download_csv(local_path: str, url: str) -> pd.DataFrame:
    os.makedirs(os.path.dirname(local_path), exist_ok=True)
    if os.path.exists(local_path):
        print(f"üîÑ √éncƒÉrcare localƒÉ: {local_path}")
        return pd.read_csv(local_path)
    else:
        print(f"üåê DescƒÉrcare de la: {url}")
        df = pd.read_csv(url)
        df.to_csv(local_path, index=False)
        print(f"‚úÖ Salvat local √Æn: {local_path}")
        return df
```

Construirea modelului de bazƒÉ

```python
model = Prophet()
model.fit(df)
```

Generarea predic»õiei (default: 30 zile)

```python
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)
```
